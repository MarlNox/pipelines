```python
import json
import sqlite3
import requests

# Define the paths to the databases
db_path_chroma = './data/vector_db/chroma.sqlite3'
db_path_webui = './data/webui.db'

# Define the base URL of your FastAPI service
base_url = "http://localhost:8080/rag/api/v1"

# Define the bearer token for authentication
bearer_token = "sk-5b5d484f375b410abc6e8b9bd46d324a"

def extract_data_to_json(db_path_chroma, db_path_webui):
    # Connect to the chroma database
    conn_chroma = sqlite3.connect(db_path_chroma)
    cursor_chroma = conn_chroma.cursor()
    
    # Connect to the webui database
    conn_webui = sqlite3.connect(db_path_webui)
    cursor_webui = conn_webui.cursor()
    
    # SQL queries to extract data
    query_collections_chroma = """
    SELECT id, name
    FROM collections
    """
    
    query_document_webui = """
    SELECT id, collection_name, filename, title, name
    FROM document
    """
    
    # Execute the queries
    cursor_chroma.execute(query_collections_chroma)
    collections_chroma = cursor_chroma.fetchall()
    
    cursor_webui.execute(query_document_webui)
    documents_webui = cursor_webui.fetchall()
    
    # Close the connections
    conn_chroma.close()
    conn_webui.close()
    
    # Process the data into the desired format
    data = {}
    
    collection_map = {collection[1]: collection[0] for collection in collections_chroma}
    
    for doc in documents_webui:
        file_id, collection_name, filename, title, name = doc
        if filename not in data:
            data[filename] = {
                "filepath": filename,
                "collections": set(),
                "metadata": {
                    "title": title,
                    "name": name
                }
            }
        data[filename]["collections"].add(collection_name)
    
    # Convert sets to lists for JSON serialization
    for filename in data:
        data[filename]["collections"] = list(data[filename]["collections"])
    
    # Convert the data dictionary to a JSON string
    json_data = json.dumps(data, indent=4)
    
    return json_data

def query_collection(collection_names, query):
    url = f"{base_url}/query/collection"
    headers = {
        "Authorization": f"Bearer {bearer_token}",
        "Content-Type": "application/json"
    }
    payload = {
        "collection_names": collection_names,
        "query": query,
        "k": 0,
        "r": 0,
        "hybrid": False
    }
    response = requests.post(url, json=payload, headers=headers)
    if response.status_code == 401:
        # Handle token expiration
        return {"error": "Token expired or invalid. Please refresh the token."}
    elif response.status_code == 200:
        return response.json()
    else:
        return {"error": f"Failed to query collections. HTTP Status Code: {response.status_code}, Detail: {response.text}"}

def get_collections_from_filenames(json_data, filenames):
    data = json.loads(json_data)
    collection_names = set()
    
    for filename in filenames:
        if filename in data:
            collection_names.update(data[filename]["collections"])
    
    return list(collection_names)

# Generate the JSON data from the databases
json_result = extract_data_to_json(db_path_chroma, db_path_webui)
data = json.loads(json_result)


# Function to extract filenames from search
def extract_filenames_from_search(search_string, collection_names):
    collection_result = query_collection(collection_names, search_string)
    if "error" in collection_result:
        return collection_result
    
    found_filenames = set()
    for meta in collection_result.get("metadatas", [[]])[0]:
        source = meta['source']
        filename = source.split('/')[-1]
        found_filenames.add(filename)
    
    return list(found_filenames)

# Function to extract content from search
def extract_content_from_search(search_string, collection_names):
    collection_result = query_collection(collection_names, search_string)
    if "error" in collection_result:
        return collection_result
    
    results = []
    for doc, meta in zip(collection_result.get("documents", [[]])[0], collection_result.get("metadatas", [[]])[0]):
        results.append(f"{meta['source']} - Page {meta['page']}: {doc}")
    
    return results



```

```python
from typing import List, Union, Generator, Iterator
import requests
import json

class Pipeline:
    def __init__(self):
        self.name = "OpenAI Pipeline"
        self.OPENAI_API_KEY = "sk-proj-bnJ4zMWclrXQIespYduAT3BlbkFJ5g95B5hUbhqPIsM6saKh"
        self.MODEL = "gpt-3.5-turbo"

    async def on_startup(self):
        print(f"on_startup:{__name__}")
        pass

    async def on_shutdown(self):
        print(f"on_shutdown:{__name__}")
        pass

    def classify_query(self, user_message: str) -> str:
        # Directional prompt for classifying the query
        classification_prompt = f"""
        Classify the following user message into one of the following categories:
        1. Information Retrieval
        2. Clarification Needed
        3. Feedback and Refinement
        4. Specific Document Retrieval
        5. Multi-step Process
        6. Normal Answer without Tool
        7. Title Generation

        User message: "{user_message}"
        """
        classification_response = self.query_openai_api([{"role": "user", "content": classification_prompt}])
        return classification_response['choices'][0]['message']['content']

    def query_openai_api(self, messages: List[dict], stream: bool = False) -> Union[dict, Generator]:
        headers = {
            "Authorization": f"Bearer {self.OPENAI_API_KEY}",
            "Content-Type": "application/json"
        }
        payload = {
            "model": self.MODEL,
            "messages": messages,
            "stream": stream
        }

        try:
            response = requests.post(
                url="https://api.openai.com/v1/chat/completions",
                json=payload,
                headers=headers,
                stream=stream
            )
            response.raise_for_status()
            if stream:
                return response.iter_lines()
            else:
                return response.json()
        except requests.exceptions.RequestException as e:
            print(f"Request failed: {e}")
            return f"Error: {e}"

    def handle_query(self, user_message: str, chat_id: str, conversation_history: List[dict]) -> str:
        classification = self.classify_query(user_message)
        response_message = ""

        if classification == "Information Retrieval":
            response_message = self.information_retrieval(user_message, conversation_history)
        elif classification == "Clarification Needed":
            response_message = self.ask_for_clarification(user_message)
        elif classification == "Feedback and Refinement":
            response_message = self.feedback_and_refinement(user_message, conversation_history)
        elif classification == "Specific Document Retrieval":
            response_message = self.specific_document_retrieval(user_message)
        elif classification == "Multi-step Process":
            response_message = self.multi_step_process(user_message, conversation_history)
        elif classification == "Normal Answer without Tool":
            response_message = self.normal_answer_without_tool(user_message)
        elif classification == "Title Generation":
            response_message = self.title_generation(user_message)
        
        return response_message

    def information_retrieval(self, user_message: str, conversation_history: List[dict]) -> str:
        # Example implementation for information retrieval
        search_prompt = f"Retrieve information based on the user's query: {user_message}"
        response = self.query_openai_api([{"role": "user", "content": search_prompt}])
        return response['choices'][0]['message']['content']

    def ask_for_clarification(self, user_message: str) -> str:
        clarification_prompt = f"Can you please provide more details about what you are looking for?"
        return clarification_prompt

    def feedback_and_refinement(self, user_message: str, conversation_history: List[dict]) -> str:
        refinement_prompt = f"Based on your previous feedback, here are more details: {user_message}. Please let me know if you need further information."
        response = self.query_openai_api([{"role": "user", "content": refinement_prompt}])
        return response['choices'][0]['message']['content']

    def specific_document_retrieval(self, user_message: str) -> str:
        doc_retrieval_prompt = f"Retrieve the document specified by the user: {user_message}"
        response = self.query_openai_api([{"role": "user", "content": doc_retrieval_prompt}])
        return response['choices'][0]['message']['content']

    def multi_step_process(self, user_message: str, conversation_history: List[dict]) -> str:
        multi_step_prompt = f"Break down the user's query into sub-queries and process each step: {user_message}"
        response = self.query_openai_api([{"role": "user", "content": multi_step_prompt}])
        return response['choices'][0]['message']['content']

    def normal_answer_without_tool(self, user_message: str) -> str:
        normal_answer_prompt = f"Provide a direct answer to the user's query without using any tools: {user_message}"
        response = self.query_openai_api([{"role": "user", "content": normal_answer_prompt}])
        return response['choices'][0]['message']['content']

    def title_generation(self, user_message: str) -> str:
        title_generation_prompt = f"Generate a title based on the user's query: {user_message}"
        response = self.query_openai_api([{"role": "user", "content": title_generation_prompt}])
        return response['choices'][0]['message']['content']

    def pipe(self, user_message: str, chat_id: str, messages: List[dict], body: dict) -> Union[str, Generator, Iterator]:
        conversation_history = messages
        user_response = self.handle_query(user_message, chat_id, conversation_history)
        conversation_history.append({"role": "assistant", "content": user_response})
        
        if 'stream' in body:
            return self.query_openai_api(conversation_history, stream=True)
        else:
            return user_response

# Example usage
pipeline = Pipeline()
print(pipeline.pipe("What is the weather today?", "chat_id_123", [{"role": "user", "content": "Hello"}], {}))
```
